***Mental Health in Tech Survey***

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/survey.csv")
df.head()
```

```python
df.columns=df.columns.str.strip().str.replace(" ","_").str.lower()
df.head()
```

```python
df.columns
```

```python
cols_to_drop=['comments','state','timestamp']
df.drop(columns=cols_to_drop,inplace=True)
df.head()
```

```python
df['gender'].unique()
```

```python
df['gender'].nunique()
```

```python

def simplify_gender(g):
    g = g.lower().strip()
    if 'female' in g:
        return 'female'
    elif 'male' in g:
        return 'male'
    elif 'm' in g:
        return 'male'
    elif 'f' in g:
        return 'female'
    elif 'Trans-female' in g:
        return 'trans'
    elif 'Trans' in g:
        return 'trans'
    elif 'non-binary' in g or 'nonbinary' in g or 'nb' in g:
        return 'non-binary'
    else:
        return 'other'
df['gender'] = df['gender'].apply(simplify_gender)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.histplot(data=df, x='age', bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
df = df[(df['age'] > 15) & (df['age'] < 65)]
```

```python
sns.histplot(data=df, x='age', bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(data=df, x='treatment', hue='gender')
plt.title('Mental Health Treatment by Gender')
plt.show()
```

```python
sns.countplot(data=df, x='family_history', hue='treatment')
plt.title('Family History vs. Treatment Seeking')
plt.show()
```

```python
binary_df = df.copy()
binary_df['treatment'] = binary_df['treatment'].map({'Yes': 1, 'No': 0})
binary_df['family_history'] = binary_df['family_history'].map({'Yes': 1, 'No': 0})
binary_df['remote_work'] = binary_df['remote_work'].map({'Yes': 1, 'No': 0})
binary_df['tech_company'] = binary_df['tech_company'].map({'Yes': 1, 'No': 0})
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(binary_df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
